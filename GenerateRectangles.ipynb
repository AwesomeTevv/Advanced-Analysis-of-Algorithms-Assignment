```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import random
```

In this program, we will be generating a list of non-overlapping, orthogonal rectangles as input for our _Brute Force_ program.

The input we generate will be in the following format:

$$
\textrm{Number},x_1,y_1,x_2,y_2
$$

where

- $\textrm{Number}$ is the rectangle number
- $(x_1, y_1)$ is the coordinates of the bottom-left corner of the rectangle
- $(x_2, y_2)$ is the coordinates of the top-right corner of the rectangle

![Rectangle Demo](https://github.com/AwesomeTevv/Advanced-Analysis-of-Algorithms-Assignment/blob/main/Images/demo-rectangle.png)


```python
max_iter = 1000

for m in range(1, max_iter):
    # Define outer rectangle
    xb = 1
    yb = 1
    xt = 20000
    yt = 20000

    x = np.array([xb, xt, xt, xb, xb])
    y = np.array([yb, yb, yt, yt, yb])

    plt.plot(x, y)

    rectangles = []

    rectangles.append([xb, yb, xt, yt])

    # Split the first 4 rectangle into 4 rectangles each
    for k in range(5):
        new = rectangles.pop(0)

        xb = new[0]
        yb = new[1]
        xt = new[2]
        yt = new[3]

        if ((xt-xb) > 1000) and ((yt - yb) > 1000):
            midx = int((xt - xb)/2) + xb
            midy = int((yt - yb)/2) + yb

            rx = random.randint(midx-400, midx+400)
            ry = random.randint(midy-400, midy+400)

            rectangles.append([xb, yb, rx, ry])
            rectangles.append([xb, ry, rx, yt])
            rectangles.append([rx, yb, xt, ry])
            rectangles.append([rx, ry, xt, yt])

            for i in range(len(rectangles)):
                xb = rectangles[i][0]
                yb = rectangles[i][1]
                xt = rectangles[i][2]
                yt = rectangles[i][3]

                x1 = np.array([xb, xt, xt, xb, xb])
                y1 = np.array([yb, yb, yt, yt, yb])

                plt.plot(x1, y1)

    # Choose a random rectange and split it into 4 rectangles
    while len(rectangles) <= m:
        thisRect = random.randint(0, len(rectangles)-1)

        new = rectangles.pop(thisRect)

        xb = new[0]
        yb = new[1]
        xt = new[2]
        yt = new[3]

        if ((xt - xb) > 1000) and ((yt - yb) > 1000):
            midx = int((xt - xb)/2) + xb
            midy = int((yt - yb)/2) + yb

            rx = random.randint(midx-200, midx+200)
            ry = random.randint(midy-200, midy+200)

            rectangles.append([xb, yb, rx, ry])
            rectangles.append([xb, ry, rx, yt])
            rectangles.append([rx, yb, xt, ry])
            rectangles.append([rx, ry, xt, yt])
        else:
            rectangles.append(new)

    # the final rectangles
    # print("The Final List of Rectangles")
    # print("------------------------------")
    for i in range(len(rectangles)):  # Plotting the current rectangles
        xb = rectangles[i][0]
        yb = rectangles[i][1]
        xt = rectangles[i][2]
        yt = rectangles[i][3]
        # print(f"{i},{xb},{yb},{xt},{yt}")
        x1 = np.array([xb, xt, xt, xb, xb])
        y1 = np.array([yb, yb, yt, yt, yb])

        plt.plot(x1, y1)

    # plt.show()
    print(f"Iteration {m}: {len(rectangles)} rectangles")
```
