```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import random
```

In this program, we will be generating a list of non-overlapping, orthogonal rectangles as input for our _Brute Force_ program.

The input we generate will be in the following format:

$$
\textrm{Number},x_1,y_1,x_2,y_2
$$

where

- $\textrm{Number}$ is the rectangle number
- $(x_1, y_1)$ is the coordinates of the bottom-left corner of the rectangle
- $(x_2, y_2)$ is the coordinates of the top-right corner of the rectangle


```python
# Define outer rectangle
xb = 1
yb = 1
xt = 20000
yt = 20000

x = np.array([xb, xt, xt, xb, xb])
y = np.array([yb, yb, yt, yt, yb])

plt.plot(x, y)

rect = []

rect.append([xb, yb, xt, yt])

# Split the first 4 rectangle into 4 rectangles each
for k in range(5):
    new = rect.pop(0)

    xb = new[0]
    yb = new[1]
    xt = new[2]
    yt = new[3]

    if ((xt-xb) > 1000) and ((yt - yb) > 1000):
        midx = int((xt - xb)/2) + xb
        midy = int((yt - yb)/2) + yb

        rx = random.randint(midx-400, midx+400)
        ry = random.randint(midy-400, midy+400)

        rect.append([xb, yb, rx, ry])
        rect.append([xb, ry, rx, yt])
        rect.append([rx, yb, xt, ry])
        rect.append([rx, ry, xt, yt])

        for i in range(len(rect)):
            xb = rect[i][0]
            yb = rect[i][1]
            xt = rect[i][2]
            yt = rect[i][3]

            x1 = np.array([xb, xt, xt, xb, xb])
            y1 = np.array([yb, yb, yt, yt, yb])

            plt.plot(x1, y1)

# Choose a random rectange and split it into 4 rectangles
while len(rect) != 100:
    thisRect = random.randint(0, len(rect)-1)

    new = rect.pop(thisRect)

    xb = new[0]
    yb = new[1]
    xt = new[2]
    yt = new[3]

    if ((xt - xb) > 1000) and ((yt - yb) > 1000):
        midx = int((xt - xb)/2) + xb
        midy = int((yt - yb)/2) + yb

        rx = random.randint(midx-200, midx+200)
        ry = random.randint(midy-200, midy+200)

        rect.append([xb, yb, rx, ry])
        rect.append([xb, ry, rx, yt])
        rect.append([rx, yb, xt, ry])
        rect.append([rx, ry, xt, yt])
    else:
        rect.append(new)

# the final rectangles
print("The Final List of Rectangles")
print("------------------------------")
for i in range(len(rect)):  # Plotting the current rectangles
    xb = rect[i][0]
    yb = rect[i][1]
    xt = rect[i][2]
    yt = rect[i][3]
    print(f"{i},{xb},{yb},{xt},{yt}")
    x1 = np.array([xb, xt, xt, xb, xb])
    y1 = np.array([yb, yb, yt, yt, yb])

    plt.plot(x1, y1)

plt.show()
```
